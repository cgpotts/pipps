This notebook just makes it easy to read the materials in different formats to make sure the examples sound okay. The tricky part of this is really making sure that the regular PP cases sound fine with both _as_ and _though_, since _as_ cannot be read concessively in this one case.

## Set-up

```python
import utils
```

```python
def read_materials(filename="materials.txt", condition="PP (No Filler/No Gap)", preposition="though", embedding=""):
    with open(filename) as f:
        materials = f.read().splitlines()
    items = [utils.item(s, preposition=preposition, embedding=embedding) for s in materials]
    for d in items:
        print(d[condition][0], end="\n\n")
```

## Though PPs

```python
read_materials(preposition="though", embedding="")
```

```python
read_materials(preposition="as", embedding="")
```

```python
read_materials(preposition="as", embedding="", condition="PiPP (Filler/Gap)")
```

```python
read_materials(preposition="asas", embedding="", condition="PiPP (Filler/Gap)")
```
